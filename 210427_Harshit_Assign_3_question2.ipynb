```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def two(img):
  clahe = cv2.createCLAHE(clipLimit=5.0)

  img = clahe.apply(img)
  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/Question2_solved_img2.JPG",img)
  return img
```

```python
def four(img):
  clahe = cv2.createCLAHE(clipLimit=8.0)
  img = clahe.apply(img)
  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/Question2_solved_img4.JPG",img)
  return img
```

```python
def three(img):
  clahe = cv2.createCLAHE(clipLimit=10.0)
  img = clahe.apply(img)
  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/Question2_solved_img3.JPG",img)
  return img
```

```python
def one(img):
  clahe = cv2.createCLAHE(clipLimit=10.0,
                          tileGridSize=(2, 2))
  img = clahe.apply(img)+5
  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/Question2_solved_img1.JPG",img)
  return img
```
