```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import math

def one(img):
  #To remove the shadows I will apply adaptive Gaussian threshold to the image:
  th = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
  cv2.THRESH_BINARY,27,11
  )

  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/solved_img1.JPG",th)
  return th
```

```python
 #similarly:
def two(img):
  th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
  cv2.THRESH_BINARY,67,9
  )

  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/solved_img2.JPG",th2)
  return th2
```

```python
def three(img):

  th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
  cv2.THRESH_BINARY,67,9
  )

  cv2.imwrite(r"/content/drive/MyDrive/Stematics/210427_Harshit_Shakya_Assignment3/solved_img3.JPG",th3)
  return th3
```

```python
if __name__=='__main__':
  img1=cv2.imread(r"/content/gutters1.JPG",cv2.IMREAD_GRAYSCALE)
  img2=cv2.imread(r"/content/gutters2.JPG",cv2.IMREAD_GRAYSCALE)
  img3=cv2.imread(r"/content/gutters3.JPG",cv2.IMREAD_GRAYSCALE)


  n=int(input("input a number (1 or 2 or 3 )"))


  if(n==1): img=one(img1)
  if(n==2): img=two(img2)
  if(n==3): img=three(img3)

  cv2_imshow(img)
```
